### Step 1 - Reading in Flat File and Reviewing Shape and Size as well as data types

```python
#Reading in data
recipe_df = pd.read_csv("RAW_Recipes.csv")

```

```python
#Checking the frame
recipe_df.head(10)

#Checking datatypes
print(recipe_df.dtypes)

#Checking shape
print("---------------------------------------")
print(recipe_df.shape)
```

### Step 2 - Trimming the dataframe
I will first get rid of columns with data we do not need for this project. The overarching directive is to have a list of recipes and ingredients that we can bump up against our other two data sources that contain nutrition and cost info. 

```python
#Removing columns we don't need. 

#Creating list of column names to decide which to drop. 
for col in recipe_df.columns:
    print(col)
    
#From the below we do not need any of the ID columns. They don't provide clear information. 
#We also don't need the year it was submitted or the description of the recipe.
#I am also going to drop the nutrition column because we are going to access our API to get nutrtion. It also isn't intuitive what this means

recipe_df = recipe_df.drop(columns=['id','contributor_id','submitted','description','nutrition'])

#Checking the columns were dropped.
recipe_df.head(5)
```

### Step 3 - Checking and removing  NAN values

```python
#Checking for NAN
print(recipe_df.isnull().sum())

#There is 1 missing value and it is in name the column. I am going to remove this. 
recipe_df = recipe_df.dropna(subset = ['name'])
```

### Step 4 - Only keeping breakfast, dinner, and lunch recipes
In the tag column, I am going to find breakfast, lunch and dinner keywords and only keep those to make my list smaller

```python
#Creating tag with boolean response to breakfast lunch and dinner
#Each is searching through the list of tags for the key word I indicate in the lambda function
recipe_df['breakfast'] = recipe_df.apply(lambda x: 1 if "breakfast" in x["tags"] else 0, axis=1)
recipe_df['lunch'] = recipe_df.apply(lambda x: 1 if "lunch" in x["tags"] else 0, axis=1)
recipe_df['dinner'] = recipe_df.apply(lambda x: 1 if "dinner" in x["tags"] else 0, axis=1)

#combining values of all three columns
recipe_df['combine'] = recipe_df.breakfast+recipe_df.lunch+recipe_df.dinner

```

```python
#combined column tells us if  the recipe is breakfast, lunch, or dinner. If it is, the value is > 1, otherwise it is <1
#Using this combined column to drop recipes not tagged breakfast, lunch or dinner

recipe_df = recipe_df[recipe_df['combine'] > 0]

recipe_df.head(5)

print(recipe_df.shape)
```

```python
#dropping the combine row, now that we no longer need it. 
recipe_df = recipe_df.drop(columns=['combine'])
```

### Step 5 - Removing recipes that are overly complex
I want this tool to be quick and easy breakfast lunch and dinner recipes. I'm going to remove recipes that have more than 10 ingredients and recipes that will take over an hour to complete

```python
#removing recipes that take more than 60 minutes
recipe_df = recipe_df[recipe_df['minutes'] <= 60]

#removing recipes that have more than 10 ingredients
recipe_df = recipe_df[recipe_df['n_ingredients'] <= 10]
```

### Step 5 - Checking for and Removing Duplicates
I will check for duplicate recipes in a few different ways. First I am going to just check for exact duplicates in the name column. Then I am going to remove stop words,whitespace, and tokenize the words from the name of the recipe and then conduct another check on duplicates. 

```python
#Checking for duplicates
numdups=recipe_df.name.duplicated().sum()
print("Number of Duplicate Recipes",numdups)

```

```python
#Removing Duplicate Recipes 
recipe_df = recipe_df.drop_duplicates(subset=['name'])

#Checking new shape
print(recipe_df.shape)

recipe_df.head(5)
```

```python
#converting all string columns to lower case (applying this to the whole frame in case we need it formatted this way for future work)
recipe_df = recipe_df.applymap(lambda s:s.lower() if type(s) == str else s)

#Removing white space from all string columns ((applying this to the whole frame in case we need it formatted this way for future work))
recipe_df = recipe_df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
#removing special characters, removing stop words

#removing special characters
recipe_df['name']=recipe_df.name.replace('\*', regex=True)

#Tokenizing to remove stop words. I need tokens otherwise it takes the letters out of other words instead of just removing stop words
from nltk.tokenize import word_tokenize

recipe_df['name_tokens'] = recipe_df['name'].apply(word_tokenize)

recipe_df.head(5)


#Removing stop words
from nltk.corpus import stopwords
import nltk

#downloading stopwords
nltk.download("stopwords")

#Creating stop words object
stop_words = stopwords.words("english")

#Removing stop words
recipe_df['name_nostop'] = recipe_df['name_tokens'].apply(lambda x: [word for word in x if word not in stop_words])

#Sorting lists of name elements
recipe_df['name_nostop'] = recipe_df.name_nostop.sort_values().apply(lambda x: sorted(x))

#Unpacking list
recipe_df['name_string'] = recipe_df['name_nostop'].apply(lambda word: ' '.join(word))
```

```python
#I have tokenized the words to remove the stop words, then I unpacked the list of names into a string again that has no stop words and is in alphabetical order
#I had to unpack because I couldn't use duplicate on list object
#Final check for duplicate recipes

numdups1=recipe_df.name_string.duplicated().sum()
print("Number of Duplicate Recipes",numdups1)

#There are an additional 328 duplicate recipes we didn't catch before. I am going to remove those. 

recipe_df = recipe_df.drop_duplicates(subset=['name_string'])

recipe_df.shape

```

#### Step 1 - Accessing the API
The API I am using has a library associated with it that makes it easy to pull the data. I am first going to import the package and create a wrapper object for my API pull

1. Source for using the Nutritionix Library - https://github.com/leetrout/python-nutritionix/

```python
#importing library. The nutrionix API has an associated library
import requests
import json
from nutritionix import Nutritionix

```

```python
def get_keys(path):
    with open(path) as f:
        return json.load(f)

keys = get_keys("api_nutrition.json")
api_key = keys['API_Key']
app_id = keys['APP_ID']

```

```python
#Creating a wrapper object for API calls
nix = Nutritionix(app_id=app_id, api_key=api_key)
```

#### Step 2  - Creating functions to pull API data
I will be using the API to pull nutrition information about the ingredients in our recipes. First I will be creating a list of ingredients to loop through and parse through the JSON repsonse data.

```python
#Creating a list of ingredients from our price dataframe
ingredient_list = price_df["ingredient"].tolist()

#Checking list
for x in ingredient_list:
    print(x)

print("-"*50)
print("There are {} ingredients in our list".format(len(ingredient_list)))
```

```python
#Checking how library functions work
#Instantiating the wrapper object and using search functionality on item
#Limiting the results to the first response only. Otherwise, the API will pull all items associated with flour
flour = nix.search("flour",results="0:1").json()

flour.keys()
print(type(flour))

for key,value in flour.items():
    print("Key : {} , Value : {}".format(key,value))

flour1= flour["hits"]
#The search option pulls basic info about the item. In order to pull nutrition information we have to search on the item_id
```

```python
#Checking formatting of lines and testing to see if we can extract the item id
print(type(flour1))
res= {}
for line in flour1:
   res.update(line)
res['_id']
```

```python
type(res)
```

```python
flour1[0]
```

```python
#Using the flour item id to pull to what info it pulls
nutrition = nix.item(id="513fceb775b8dbbc21002de8").json()
nutrition
```

```python
nutrition["item_name"]
```

```python
#First I want to parse through the list and pull the item_id out of each search item
#The item id is nested so I'm initiating two objects. First, am empty list to put the item ID's into.
#Second is a new dictionary object that takes the nested items under the 'hit' key and then accessing the id within it
item_id_list = []
res= {}

for item in ingredient_list:
    json_data = nix.search(item, results="0:1").json() #pulling data for eac item in json format
    key1= json_data["hits"]
    for line in key1:
       res.update(line) #adding nested dictionary to it's own dictionary, then accessing each item of the new dictionary
    item_id = res['_id']
    item_id_list.append(item_id)
    
    
```

```python
#Checking the list
print(item_id_list)

print(len(item_id_list))
```

#### Step 3 Accessing nutrtion information and creating a dataframe
In order to access the nutrition information, I had to extract the item ID for each item. Now we will use that item ID to pull in the nutrition info.

```python
#Creating empty list objects
name = []
calories = []
fat = []
protein = []
carbs = []
serving_size = []
serving_unit = []
serving_weight = []

#Using keys to append details to list
for item in item_id_list:
    json_data = nix.item(id =item).json()
    name.append(json_data['item_name'])
    calories.append(json_data['nf_calories'])
    fat.append(json_data['nf_total_fat'])
    protein.append(json_data['nf_protein'])
    carbs.append(json_data['nf_total_carbohydrate'])
    serving_size.append(json_data['nf_serving_size_qty'])
    serving_unit.append(json_data['nf_serving_size_unit'])
    serving_weight.append(json_data['nf_serving_weight_grams'])
    
    
    

```

```python
#Creating a dataframe from the lists
nutrition_df = pd.DataFrame({'ingredient':name,'calories':calories,'fat':fat,'protein':protein,'carbs':carbs,'serving_size':serving_size
                             ,'serving_unit':serving_unit,'serving_weight':serving_weight})
```

#### Step 3 Creating database and loading tables

```python
#importing necessary libraries
import sqlite3
```

```python
#Creating the database and a cursor object
conn = sqlite3.connect('recipe_database')
c = conn.cursor()
```

```python
#Creating table for recipes_df
c.execute('CREATE TABLE IF NOT EXISTS recipes (minutes number, tags text, n_steps number,steps text,ingredients text,n_ingredients number,breakfast number,lunch number, dinner number, name_string text, contains_rice text)')
conn.commit()
```

```python
#Creating a table for the nutrition_df
c.execute('CREATE TABLE IF NOT EXISTS nutrition (calories number,fat number,protein number,carbs number, serving_size number,serving_unit text,serving_weight number, item text)')
conn.commit()
```

```python
#Creating a table for the price dataframe
c.execute('CREATE TABLE IF NOT EXISTS prices (ingredient text, price number, unit_grams number)')
conn.commit()
```

```python
#loading recipe data into SQL recipe table
recipe_df.to_sql('recipes',conn, if_exists = 'replace',index = False)
```

```python
#loading nutrition dataframe in SQL nutrition table
nutrition_df.to_sql('nutrition',conn, if_exists = 'replace', index = False)
```

```python
#loading price data into SQL price table
price_df.to_sql('prices', conn, if_exists = 'replace', index = False)
```

In order to merge our database tables, there must be an identifier that is common among the three table. For this project, the ingredient would be the common identifier. However there is a complication. For each recipe, the ingredients are contained in a list object for each recipe. In order to demonstrate merging the three tables, I am going to use 

```python
#performing an inner join on the name of the ingredient and putting results into rows object
data = []
sql = """SELECT * FROM prices LEFT JOIN nutrition ON prices.ingredient = nutrition.item LEFT JOIN recipes ON recipes.contains_rice = nutrition.item"""
rows = c.execute(sql)
for row in rows:
    data.append(row)
```
